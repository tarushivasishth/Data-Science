```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import load_digits
```

```python
dataset = load_digits()
dataset
```

```python
dataset["data"]
# pixel values
```

```python
dataset.data.shape
```

```python
df = pd.DataFrame(dataset.data,columns=dataset.feature_names)
df["target"] = dataset.target
df
```

```python
plt.gray()
plt.matshow(dataset.data[1].reshape(8,8))
```

```python
X = dataset.data
y = dataset.target
```

```python
# scaling the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(X)
X_scaled = scaler.transform(X)
X,X_scaled
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X_scaled,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
```

```python
clf.fit(xtrain,ytrain)
```

```python
from sklearn.metrics import accuracy_score
y_train_pred = clf.predict(xtrain)
y_test_pred = clf.predict(xtest)
```

```python
accuracy_score(ytrain,y_train_pred)
```

```python
accuracy_score(ytest,y_test_pred)
```

<h3>PCA - either use (1) n_components or (2) provide with the %age of important features that should be retained</h3>

```python
# 2: using %
from sklearn.decomposition import PCA
pca = PCA(0.95) # 95% of important features are retained
```

```python
X_pca = pca.fit_transform(X_scaled)
X_pca
```

```python
X_pca.shape,X.shape
```

```python
pca.explained_variance_ratio_ # % of important features retained from each column
```

```python
pca.n_components_
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X_pca,y,test_size=0.2,random_state=42)
```

```python
clf = LogisticRegression(max_iter=1000)
clf.fit(xtrain,ytrain)
```

```python
y_train_pred = clf.predict(xtrain)
y_test_pred = clf.predict(xtest)
```

```python
accuracy_score(ytrain,y_train_pred)
```

```python
accuracy_score(ytest,y_test_pred)
```
