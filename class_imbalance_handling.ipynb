```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from keras.layers import Dense,Dropout
from keras.models import Sequential
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv(r"./datasets/Churn_Modelling.csv")
df.head()
```

```python
df.info()
# no missing value
```

```python
df.shape
```

```python
df.drop(["RowNumber",'CustomerId',"Surname"],axis=1,inplace=True)
```

```python
df
```

```python
for i in df.columns:
    print(f"{i}: {df[i].unique()}")
```

```python
# encoding
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df["Gender"] = encoder.fit_transform(df["Gender"])
df["Geography"] = encoder.fit_transform(df["Geography"])
```

```python
for i in df.columns:
    print(f"{i}: {df[i].unique()}")
```

```python
# scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
numnames = ["EstimatedSalary","Balance","Age","CreditScore","Tenure"]
```

```python
df[numnames] = scaler.fit_transform(df[numnames])
```

```python
for i in df.columns:
    print(f"{i}: {df[i].unique()}")
```

```python
X = df.drop(["Exited"],axis=1)
y = df["Exited"]
```

```python
Xtrain,Xtest,ytrain,ytest = train_test_split(X,y,test_size=0.20,random_state=5)
```

```python
ytrain.value_counts()
```

```python
def model_training_ANN(Xtrain,Xtest,ytrain,ytest,epochs):
    model = Sequential([
        Dense(10,input_shape=(10,),activation="relu"),
        Dense(20,activation="relu"),
        Dense(1,activation="sigmoid"),
    ])

    model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])

    model.fit(Xtrain,ytrain,epochs=epochs)
    ypred = np.round(model.predict(Xtest))
    print(classification_report(ytest,ypred))
    return ypred
```

```python
model_training_ANN(Xtrain,Xtest,ytrain,ytest,100)
```

# Handling imbalance in the data

```python
df.Exited.value_counts()
```

# random undersampling

```python
df_class0 = df[df.Exited == 0]
df_class1 = df[df.Exited == 1]
```

```python
count_class0 = len(df_class0)
count_class1 = len(df_class1)
```

```python
df1 = pd.concat([df_class0[:count_class1],df_class1],axis=0)
```

```python
df1.Exited.value_counts()
```

```python
X = df1.drop(["Exited"],axis=1)
y = df1["Exited"]
```

```python
Xtrain,Xtest,ytrain,ytest = train_test_split(X,y,test_size=0.20,random_state=5,stratify=y)
```

```python
model_training_ANN(Xtrain,Xtest,ytrain,ytest,100)
```

# random oversampling

```python
df.Exited.value_counts()
```

```python
df2 = pd.concat([df_class0,df_class1.sample(count_class0,replace=True)],axis=0)
```

```python
df2.Exited.value_counts()
```

```python
X = df2.drop(["Exited"],axis=1)
y = df2["Exited"]
```

```python
Xtrain,Xtest,ytrain,ytest = train_test_split(X,y,test_size=0.20,random_state=5,stratify=y)
```

```python
model_training_ANN(Xtrain,Xtest,ytrain,ytest,100)
```

# oversampling using smote

```python
X = df.drop(["Exited"],axis=1)
y = df.Exited
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(sampling_strategy="minority")
X_sm,y_sm = smote.fit_resample(X,y)
```

```python
y_sm.value_counts()
```

```python
Xtrain,Xtest,ytrain,ytest = train_test_split(X_sm,y_sm,test_size=0.20,random_state=5,stratify=y_sm)
```

```python
model_training_ANN(Xtrain,Xtest,ytrain,ytest,100)
```

# using ensemble

```python
df.Exited.value_counts()
```

```python
7963/2037
```

```python
Xtrain,Xtest,ytrain,ytest = train_test_split(X,y,test_size=0.20,random_state=5,stratify=y)
```

```python
ytrain.value_counts()
```

```python
df3 = Xtrain.copy()
df3["Exited"] = ytrain
```

```python
df3_class0 = df3[df3["Exited"] == 0]
df3_class1 = df3[df3["Exited"] == 1]
```

```python
def train_ensemble_data(df_majority,df_minority,start,end):
    df_train = pd.concat([df_majority[start:end],df_minority],axis=0)
    Xtrain = df_train.drop(["Exited"],axis=1)
    ytrain = df_train["Exited"]
    return Xtrain,ytrain
```

```python
Xtrain,ytrain = train_ensemble_data(df3_class0,df3_class1,0,1630)
```

```python
ypred1 = model_training_ANN(Xtrain,Xtest,ytrain,ytest,100)
```

```python
Xtrain,ytrain = train_ensemble_data(df3_class0,df3_class1,1630,3260)
```

```python
ypred2 = model_training_ANN(Xtrain,Xtest,ytrain,ytest,100)
```

```python
Xtrain,ytrain = train_ensemble_data(df3_class0,df3_class1,3260,6370)
```

```python
ypred3 = model_training_ANN(Xtrain,Xtest,ytrain,ytest,100)
```

```python
ypred1
```

```python
y_final_pred = ypred1.copy()
for i in range(len(ypred1)):
    n_ones = ypred1[i] + ypred2[i] + ypred3[i]
    if n_ones > 1:
        y_final_pred[i] = 1
    else:
        y_final_pred[i] = 0
```

```python
print(classification_report(ytest,y_final_pred))
```

<h4>Therefore, we obtained the best results using SMOTE oversampling technique</h4>
