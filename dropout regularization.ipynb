```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense,Dropout

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r"C:\Users\DeveshNarain\deep_learning_codebasics\datasets\sonar_dataset.csv",header=None)
df.sample(5)
```

```python
df.columns
# col 60 is the target column
```

```python
df.info()
```

```python
df[60].value_counts()
```

```python
df1 = pd.get_dummies(df,columns=[60],drop_first=True)
```

```python
df1
# 1 -> R and 0 -> M
```

```python
X = df1.drop(['60_R'],axis=1)
y = df1['60_R']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,ytrain,ytest = train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
Xtrain.shape
```

```python
Xtest.shape
```

# Model Building without Dropout layer

```python
model = Sequential([
    Dense(60,input_shape=(60,),activation='relu'),
    Dense(30,activation='relu'),
    Dense(15,activation='relu'),
    Dense(1,activation='sigmoid'),
])

model.compile(optimizer="adam",
              loss="binary_crossentropy",
              metrics=['accuracy']
             )

model.fit(Xtrain,ytrain,epochs=100,batch_size=8) # mini-batch gradient descent
```

```python
model.evaluate(Xtest,ytest) 
# overfitting
```

```python
ypred = model.predict(Xtest)
ypred = np.round(ypred)
ypred[:10]
```

```python
ytest[:10]
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,ypred))
```

# Model with Dropout layer in between hidden layers

```python
model_dropout = Sequential([
    Dense(60,input_shape=(60,),activation='relu'),
    Dropout(0.5),
    Dense(30,activation='relu'),
    Dropout(0.5),
    Dense(30,activation='relu'),
    Dropout(0.5),
    Dense(1,activation='sigmoid'),
])

model_dropout.compile(optimizer="adam",
              loss="binary_crossentropy",
              metrics=['accuracy']
             )

model_dropout.fit(Xtrain,ytrain,epochs=100,batch_size=8) # mini-batch gradient descent
```

```python
model_dropout.evaluate(Xtest,ytest)
```

```python
ypred_drop = model_dropout.predict(Xtest)
ypred_drop = np.round(ypred)
ypred_drop[:10]
```

```python
ytest[:10]
```

```python
print(classification_report(ytest,ypred_drop))
```
