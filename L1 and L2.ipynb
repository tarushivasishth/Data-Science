```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv(r"D:\ML codebasics\datasets\Melbourne_housing_FULL.csv")
df
```

```python
df.shape
```

```python
df.nunique()
```

```python
cols_to_use = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount', 
               'Distance', 'CouncilArea', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']
df = df[cols_to_use]
```

```python
df.info()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
cols_with_0 = ["Car","Bathroom","Bedroom2","Propertycount","Distance"]
df[cols_with_0] = df[cols_with_0].fillna(0)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df["BuildingArea"].fillna(np.mean(df["BuildingArea"]),inplace=True)
df["Landsize"].fillna(np.mean(df["Landsize"]),inplace=True)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df.dropna(axis=0,inplace=True)
```

```python
df.info()
```

```python
df.shape
```

```python
# one hot encoding
df = pd.get_dummies(df,drop_first=True) # to prevent dummy variable trap
df
```

# Model Building and Training

```python
x = df.drop("Price",axis=1)
y = df["Price"]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
model.score(xtrain,ytrain)
```

# L1 regularization

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso = GridSearchCV(Lasso(),{
    "alpha": [20,30,50],
    "max_iter": [10,50,100],
    "tol": [0.1,0.01,0.005]
},cv=5)

lasso.fit(x,y)
```

```python
lasso.best_params_
```

```python
lasso = Lasso(alpha=20,max_iter=100,tol=0.1)
```

```python
lasso.fit(xtrain,ytrain)
```

```python
lasso.score(xtest,ytest)
```

```python
lasso.score(xtrain,ytrain)
```

# L2 regularization

```python
from sklearn.linear_model import Ridge
ridge = GridSearchCV(Ridge(),{
    "alpha": [20,30,50],
    "max_iter": [10,50,100],
    "tol": [0.1,0.01,0.005]
},cv=5)

ridge.fit(x,y)
```

```python
ridge.best_params_
```

```python
ridge.best_score_
```

```python
ridge = Ridge(alpha=20,max_iter=10,tol=0.1)
```

```python
ridge.fit(xtrain,ytrain)
```

```python
ridge.score(xtest,ytest)
```

```python
ridge.score(xtrain,ytrain)
```
