```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("E:\hair_loss.csv")
```

```python
df.head()
```

```python
df.drop(['total_protein'],axis=1)
```

```python
df.info()
```

```python
df.shape
```

```python
df = df.loc[(df['total_keratine']<10) & (df['vitamin']<100) & (df['iron']<50)]
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df
```

```python
#bar plot
sns.barplot(df,x='total_keratine',y='vitamin')
plt.title("bar plot")
```

```python
#bar plot
plt.bar(df['total_keratine'],df['vitamin'])
plt.xlabel("total keratine")
plt.ylabel("vitamin")
plt.title("bar plot")
```

```python
sns.barplot(df,x='total_keratine',y='vitamin',color='yellow')
```

```python
sns.barplot(df,x='iron',y='calcium',palette='rocket')
```

```python
# countplot
sns.countplot(df,x='stress_level',palette='Set1')
```

```python
sns.countplot(df,y='hair_fall',palette='Set1',hue='hair_fall')
```

```python
sns.countplot(df,x='hair_fall',palette='Set1',hue='total_keratine')
```

```python
#distribution plot
sns.distplot(df.calcium,bins=12,color='c')
```

```python
#heatmaps
sns.heatmap(df.corr(),cbar=True,linewidth=0.5)
#corr is for correlation and cbar is for color bar
#lighter color have +ve corr while those having darker color ahve -ve corr
```
