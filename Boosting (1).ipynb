```python
le = LabelEncoder()
for i in df.columns:
  df[i] = le.fit_transform(df[i])
```

```python
df
```

```python
X = df.drop('class', axis=1)
y = df['class']
```

```python
Xtrain,Xtest,ytrain,ytest = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(criterion='entropy',max_depth=1)
AdaBoost = AdaBoostClassifier(base_estimator=model,n_estimators=400,learning_rate=1)
boostmodel = AdaBoost.fit(Xtrain,ytrain)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(boostmodel,X,y,cv=5)
scores
```

```python
scores.mean()
```

```python
from sklearn.model_selection import GridSearchCV
boostmodel = GridSearchCV(AdaBoost,{
    'n_estimators':[10,50,100,500,1000],
   'base_estimator':[DecisionTreeClassifier(criterion='entropy',max_depth=1)],
    'learning_rate':[0.01,0.1,1]
},cv=5,return_train_score=False)
boostmodel.fit(X,y)
```

```python
dir(boostmodel)
```

```python
boostmodel.best_score_
```

```python
boostmodel.score(Xtest,ytest)
```
